```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#2. Import, read and display the document
df = pd.read_csv(r"Ice Cream Ratings.csv")
df. set_index('Date')
df
```

```python
#Optional step: Check styles of charts and then apply
print(plt.style.available)
plt.style.use('dark_background')
```

```python
#Line plot
df.plot(kind = 'line')
```

```python
#Seperating curve using 'subplots'
df.plot(kind = 'line', subplots = True)
```

```python
#add titles and axis labels
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = "Daily Ratings", ylabel = 'Scores')
```

```python
#bar plot
df.plot(kind='bar')
```

```python
#stacked bar chart
df.plot(kind='bar', stacked=True, title='Ice Cream Ratings', xlabel='Flavours', ylabel='scores')
```

```python
#only one flavour
df['Flavor Rating'].plot(kind='bar', stacked=True)
```

```python
#horizontal graph
df.plot.barh(stacked=True)
```

```python
#scatter plot
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s=250, c="orange")
```

```python
#histogram
#df.plot.hist()
df.plot.hist(bins = 20)

#bins represent how many markigs/columns are on the x axis
```

```python
#boxplot

df.boxplot()
#a box plot shows max and medium
#it shows the 25th, 50th and 75th percentile to show the difference etween the skews
```

```python
#area plot

#df.plot.area()
#to icrease size

df.plot.area(figsize=(10,5))
```

```python
#pie chart
df.plot.pie(y='Flavor Rating', figsize = (10,10))
```
